````python
%%sql
SELECT property_occupied.st_property, property_occupied.occupied/365*100 AS occupancy_rate, st_price.sample_nightly_rent_price
FROM(
    SELECT st_property, COUNT(DISTINCT rental_date) AS occupied
    FROM st_rental_dates
    WHERE EXTRACT(YEAR FROM rental_date) = 2015
    GROUP BY st_property
    ) AS property_occupied
JOIN st_property_info st_info
ON property_occupied.st_property = st_info.st_property_id
JOIN watershed_property_info ws
ON st_info.location = ws.location AND st_info.property_type = ws.property_type
JOIN st_rental_prices st_price
ON st_price.location = ws.location AND st_price.property_type = ws.property_type
WHERE ws.ws_property_id = 'W24';
````

````python
%%sql
SELECT ws.ws_property_id, ws.current_monthly_rent, l.*, p.*, property_occupied.st_property, property_occupied.occupied/365*100 AS occupancy_rate, st_price.sample_nightly_rent_price, st_price.percentile_10th_price, st_price.percentile_90th_price
FROM(
    SELECT st_property, COUNT(DISTINCT rental_date) AS occupied
    FROM st_rental_dates
    WHERE EXTRACT(YEAR FROM rental_date) = 2015
    GROUP BY st_property
    ) AS property_occupied
JOIN st_property_info st_info
ON property_occupied.st_property = st_info.st_property_id
JOIN watershed_property_info ws
ON st_info.location = ws.location AND st_info.property_type = ws.property_type
JOIN st_rental_prices st_price
ON st_price.location = ws.location AND st_price.property_type = ws.property_type
JOIN property_type p
ON st_info.property_type = p.property_type_id
JOIN location l
ON st_info.location = l.location_id
LIMIT 10;
````

````python
my_data=%sql SELECT ws.ws_property_id, ws.current_monthly_rent, l.*, p.*, property_occupied.st_property, property_occupied.occupied/365*100 AS occupancy_rate, st_price.sample_nightly_rent_price, st_price.percentile_10th_price, st_price.percentile_90th_price FROM(SELECT st_property, COUNT(DISTINCT rental_date) AS occupied FROM st_rental_dates WHERE EXTRACT(YEAR FROM rental_date) = 2015 GROUP BY st_property) AS property_occupied JOIN st_property_info st_info ON property_occupied.st_property = st_info.st_property_id JOIN watershed_property_info ws ON st_info.location = ws.location AND st_info.property_type = ws.property_type JOIN st_rental_prices st_price ON st_price.location = ws.location AND st_price.property_type = ws.property_type JOIN property_type p ON st_info.property_type = p.property_type_id JOIN location l ON st_info.location = l.location_id;
````
